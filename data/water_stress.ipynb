```python
water_stress = pd.read_csv('water_stress.csv')
```

```python
water_stress_2017 = water_stress[water_stress.Year == 2017]
```

```python
water_stress_2017.rename(columns = {'Annual freshwater withdrawals, total (% of internal resources)':'percent'}, inplace = True)
```

```python
conditions = [
    (water_stress_2017['percent'] < 10),
    (water_stress_2017['percent'] < 20),
    (water_stress_2017['percent'] < 40),
    (water_stress_2017['percent'] < 80),
    (water_stress_2017['percent'] >= 80)
]

values = ['Low stress', 'Low-to-medium stress', 'Medium-to-high stress', 'High stress', 'Extremely high stress']

water_stress_2017['level'] = np.select(conditions, values)

water_stress_2017.head()
```

```python
countries = gpd.read_file('countries.geojson')
countries.head()
```

```python
waterstress = countries.merge(water_stress_2017, how='outer', left_on='ISO_A3', right_on='Code')
```

```python
waterstress = waterstress.dropna(subset=['ISO_A3'])
```

```python
len(waterstress[waterstress.level == 'Extremely high stress'])
```

```python
waterstress[waterstress.level == 'Extremely high stress']
```

```python
len(waterstress[waterstress.level == 'High stress'])
```

```python
waterstress[waterstress.level == 'High stress']
```

```python
waterstress.to_file('waterstress.geojson', driver='GeoJSON')
```
